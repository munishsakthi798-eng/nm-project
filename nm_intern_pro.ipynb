```python
import pandas as pd

# Sample data (you can replace this with your dataset)
data = {
    'Item': ['Biryani', 'Pizza', 'Burger', 'Dosa', 'Biryani', 'Pizza', 'Biryani', 'Burger', 'Dosa', 'Biryani'],
    'Hour': [12, 13, 14, 18, 19, 20, 21, 19, 20, 19],
    'Location': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C']
}

df = pd.DataFrame(data)

# 🔹 PEAK ORDER TIMES
print("\n========== PEAK ORDER TIMES ==========")
peak_orders = df['Hour'].value_counts().sort_index()
print(peak_orders)

# 🔹 MOST POPULAR DISHES
popular_items = df['Item'].value_counts()

print("\n========== MOST POPULAR DISHES ==========")
print(popular_items)

# 🔹 LOCATION WISE ORDERS (optional)
location_orders = df['Location'].value_counts()

print("\n========== LOCATION ORDERS ==========")
print(location_orders)
```

```python
print("\n========== LOCATION ANALYSIS ==========")
print(location_orders)
```

```python
plt.figure(figsize=(8,5))
peak_orders.plot(kind='bar')

plt.title('Peak Order Times')
plt.xlabel('Hour')
plt.ylabel('Number of Orders')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
popular_items.plot(kind='bar')
plt.title('Most Popular Food Items')
plt.xlabel('Food Item')
plt.ylabel('Order Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
location_orders.plot(kind='pie', autopct='%1.1f%%')
plt.title('Orders by Location')
plt.ylabel('')
plt.show()
```

```python
print("\n========== BUSINESS INSIGHTS ==========")

print("Most Popular Dish :",
      popular_items.idxmax())

print("Peak Order Hour :",
      peak_orders.idxmax())

print("Top Location :",
      location_orders.idxmax())

print("\nProject Completed Successfully!")
```
